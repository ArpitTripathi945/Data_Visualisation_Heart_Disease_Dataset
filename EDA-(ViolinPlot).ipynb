# Violin Plot for EDA

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_csv('heart.csv')
df.head()
```

```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Age'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'])
```
