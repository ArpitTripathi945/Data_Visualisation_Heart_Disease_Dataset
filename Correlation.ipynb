# CORRELATION

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_csv('heart.csv')
df.head()
```

```python
df.corr()
```

## HEAT MAP


```python
sns.heatmap(df.corr())
```

## JOINT PLOT

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df)
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'Cholesterol', y = 'FastingBS', data = df, kind = 'reg')
```
