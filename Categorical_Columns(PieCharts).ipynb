# CATEGORICAL COLUMNS - PIE CHARTS

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_csv('heart.csv')
df.head()
```

# Distribution Plot for Numeric Columns

```python
sns.distplot(df['RestingBP'], kde = 'True', color='blue')
```

```python
sns.distplot(df['Cholesterol'], kde = 'True', color='blue')
```

```python
sns.distplot(df['Oldpeak'], kde = 'True', color='blue')
```

# Pie Charts for Categorical

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f')
```
